```python
import gradio as gr
from llama_cpp import Llama

# Path to your downloaded GGUF model
model_path = "/root/.cache/huggingface/hub/models--JaaackXD--Llama-3-8B-GGUF/snapshots/06fcea7d93a11438e89d417adf7ac9d1a9ec83a2/ggml-model-Q4_K_M.gguf"

# Initialize the Llama model
llm = Llama(model_path=model_path, temperature=0.7)

# Define chatbot function
def chat_with_llama(user_input):
    if not user_input.strip():
        return "Please enter a message."

    output = llm(user_input, max_tokens=100)
    response_text = output["choices"][0]["text"].strip()

    return response_text

# Create Gradio chatbot UI
chatbot_ui = gr.Interface(
    fn=chat_with_llama,
    inputs=gr.Textbox(placeholder="Type your message here..."),
    outputs="text",
    title="🦙 Llama 3 Chatbot",
    description="Chat with an AI powered by Llama 3! Type a message below.",
    theme="default"
)

# Launch the chatbot
if __name__ == "__main__":
    chatbot_ui.launch()
```

```python
from llama_cpp import Llama

# Path to your downloaded GGUF model
model_path = "/root/.cache/huggingface/hub/models--JaaackXD--Llama-3-8B-GGUF/snapshots/06fcea7d93a11438e89d417adf7ac9d1a9ec83a2/ggml-model-Q4_K_M.gguf"

# Initialize the Llama model
llm = Llama(model_path=model_path, temperature=0.7)

# Define chatbot function
def chat_with_llama(user_input):
    if not user_input.strip():
        return "Please enter a valid message."

    prompt = f"User: {user_input}\nAssistant:"
    output = llm(prompt, max_tokens=100)

    # Extract and return response text
    response_text = output["choices"][0]["text"].strip()
    return response_text

# Simple CLI chatbot loop
print("🦙 Llama 3 Chatbot (Type 'exit' to stop)")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Goodbye!")
        break

    response = chat_with_llama(user_input)
    print(f"Llama: {response}")
```
