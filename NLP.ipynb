```python
import gradio as gr
import pandas as pd
from transformers import pipeline
from collections import Counter

# Load sentiment analysis model
sentiment_pipeline = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")

# Label mapping for Cardiff NLP sentiment model
label_mapping = {
    "LABEL_0": "Negative",
    "LABEL_1": "Neutral",
    "LABEL_2": "Positive"
}

# Function to analyze social media posts
def analyze_social_media(posts):
    sentiments = []

    for post in posts.split("\n"):  # Split input into multiple lines
        if post.strip():  # Ignore empty lines
            result = sentiment_pipeline(post)[0]
            sentiment_label = label_mapping[result['label']]  # Convert model output to human-readable labels
            sentiments.append({
                "Post": post,
                "Sentiment": sentiment_label,
                "Confidence": f"{result['score'] * 100:.2f}%"
            })

    # Count sentiment distribution
    sentiment_counts = Counter([entry["Sentiment"] for entry in sentiments])
    summary = f"Positive: {sentiment_counts.get('Positive', 0)}, Neutral: {sentiment_counts.get('Neutral', 0)}, Negative: {sentiment_counts.get('Negative', 0)}"

    # Convert to DataFrame for display
    df = pd.DataFrame(sentiments)

    return df, summary

# Gradio Interface
interface = gr.Interface(
    fn=analyze_social_media,
    inputs=gr.Textbox(lines=5, label="Enter Social Media Posts (one per line)", placeholder="Type or paste posts here..."),
    outputs=[gr.Dataframe(headers=["Post", "Sentiment", "Confidence"]), gr.Textbox(label="Overall Sentiment Summary")],
    title="Social Media Sentiment Monitoring",
    description="Analyze social media posts and classify them as Positive, Neutral, or Negative. Get a sentiment trend summary.",
)

# Launch the app
interface.launch()
```

```python
from spellchecker import SpellChecker
import language_tool_python

# Function to check spelling
def check_spelling(text):
    spell = SpellChecker()
    words = text.split()
    misspelled = spell.unknown(words)

    corrections = {}
    for word in misspelled:
        corrections[word] = spell.correction(word)

    return corrections

# Function to check grammar
def check_grammar(text):
    tool = language_tool_python.LanguageTool('en-US')
    matches = tool.check(text)
    return matches

# Input paragraph
text = input("Enter a paragraph of text for checking: ")

# Check spelling
spelling_corrections = check_spelling(text)
if spelling_corrections:
    print(f"\nSpelling corrections: {spelling_corrections}")
else:
    print("\nNo spelling errors found.")

# Check grammar
grammar_issues = check_grammar(text)
if grammar_issues:
    print(f"\nGrammar issues found: {len(grammar_issues)}")
    for issue in grammar_issues:
        print(f"Issue: {issue.ruleId}, Message: {issue.message}")
else:
    print("\nNo grammar issues found.")
```

```python
import gradio as gr
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# ✅ NEW Sample dataset (General Knowledge, Literature, Science, and History)
existing_texts = [
    "The Renaissance was a cultural movement that profoundly affected European intellectual life.",
    "Gravity is a force that attracts two bodies toward each other, as described by Isaac Newton.",
    "Shakespeare's Hamlet explores themes of madness, betrayal, and revenge.",
    "The Industrial Revolution marked a period of major industrialization from the 18th to 19th centuries.",
    "The mitochondrion is often called the powerhouse of the cell because it generates ATP.",
    "The Great Wall of China was built to protect against invasions from northern tribes.",
    "World War II lasted from 1939 to 1945 and involved many countries around the globe.",
    "The theory of relativity was developed by Albert Einstein and revolutionized physics.",
    "Leonardo da Vinci painted the Mona Lisa, one of the most famous artworks in history.",
    "The process of photosynthesis allows plants to convert sunlight into energy, producing oxygen.",
]

# Vectorizer for text similarity
vectorizer = TfidfVectorizer()

# Function to check for plagiarism
def check_plagiarism(input_text):
    texts = existing_texts + [input_text]
    tfidf_matrix = vectorizer.fit_transform(texts)

    # Compute cosine similarity
    similarity_scores = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])[0]

    # Check if similarity is too high
    max_score = max(similarity_scores)
    threshold = 0.7  # Define plagiarism threshold (70%)

    if max_score > threshold:
        return f"⚠ High similarity detected! ({max_score * 100:.2f}% match)"
    else:
        return "✅ No significant plagiarism detected."

# Gradio Interface
interface = gr.Interface(
    fn=check_plagiarism,
    inputs=gr.Textbox(lines=5, label="Enter Text", placeholder="Paste your text here..."),
    outputs=gr.Textbox(label="Plagiarism Check Result"),
    title="Plagiarism Detection System",
    description="Checks if the input text is too similar to existing content using NLP techniques.",
)

# Launch the Gradio app in Colab
interface.launch(share=True)
```
